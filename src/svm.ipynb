```python
from sklearn.model_selection import train_test_split
```

```python
resetado=resetado.astype({"ANO" : str})
```

Prevendo o consumo anual de cada região

```python

resetado = resetado.sort_values("ANO")
X = resetado[resetado["REGIAO"] == "Norte"]["ANO"].reset_index().drop("index", axis=1)
y = resetado[resetado["REGIAO"] == "Norte"]["total_por_regiao"].reset_index().drop("index", axis=1)

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, random_state=1, test_size=.25)    
```

```python
from sklearn.svm import SVR

model = SVR(kernel="linear", C=100)

model.fit(Xtrain, ytrain["total_por_regiao"])

ymodel=model.predict(Xtest)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(X["ANO"], y["total_por_regiao"])
plt.scatter(Xtrain["ANO"], ytrain["total_por_regiao"])
plt.scatter(Xtest["ANO"], ytest["total_por_regiao"],)

plt.show()
```

```python
plt.scatter(Xtest["ANO"],ytest, label="dado de treino")
plt.scatter(Xtest["ANO"], ymodel, label="modelo")
plt.xlabel("ANO")
plt.ylabel("Consumo")
plt.legend()
plt.show()
```

```python
residuais=ytest["total_por_regiao"] - ymodel
plt.title("Residuais x Valor estimado")
plt.scatter(ymodel, residuais)
plt.axhline(y=0, color="r", linestyle="--")
plt.xlabel("Consumo previsto")
plt.ylabel("Erro residual")
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(ytest, ymodel)
mse = mean_squared_error(ytest, ymodel)
r2 = r2_score(ytest, ymodel)

print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```
